## Task 1: A list of dictionaries

### a) Filter out invalid transactions where the price is zero or the quantity is negative.

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# Removes transactions where price equals zero or quantity is negative
for i in transactions[:]:
    if i["price"] == 0 or i["quantity"] < 0:
        transactions.remove(i)

# Prints the remaining transactions
print("Remaining transactions:")
for i in transactions:
    print(i)
```

### b) For each remaining transaction, compute a new field `total` = `price * quantity`.

```python
# Adds new field total (price * quantity)
for i in transactions:
    i["total"] = i["price"] * i["quantity"]
    print(i)
```

### c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

```python
revenue_per_region = {}

# Creates keys for each region in a dictionary
for i in transactions:
    region = i["region"]
    region_revenue = i["total"]
    
    # Assigns a temporary value to each key
    revenue_per_region[region] = 0

# Loops through all regions and computes the total revenue for each region
for i in transactions:
    region = i["region"]
    region_revenue = i["total"]

    revenue_per_region[region] = revenue_per_region[region] + region_revenue

print(f"Total revenue per region:\n{revenue_per_region}")
```

### d) Print the region with the highest total revenue and its value using an f-string.

```python
highest_value = 0
region_highest_value = ""

# Loops through all regions and assigns the highest values and the corresponding region to variables
for region, value in revenue_per_region.items():
    if highest_value < value:
        highest_value = value
        region_highest_value = region

print(f"Region with the highest value: {region_highest_value} ({highest_value})")
```

## Task 2: Creating a function

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(sales:list):
    total = sum(sales)

    average = round((total / len(sales)), 2) # Returns the mean daily sales value and rounds to two decimals
    
    # Loops through the list and returns the index of the day with the highest and lowest sales
    day_index = 0
    best_day_value = 0
    worst_day_value = max(sales) # Temporarily assigns the worst_day_value to the highest value in the list
    
    for day in sales:
        day_index += 1

        # Assigns the best day
        if day > best_day_value:
            best_day_value = day
            best_day = day_index

        # Assigns the worst day
        if day < worst_day_value:
            worst_day_value = day
            worst_day = day_index

    # Loops through the list and returns the amount of days where sales exceed the average
    above_average_count = 0
    for day in sales:
        if day > average:
            above_average_count += 1

    summarised_sales = {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }

    return summarised_sales

print(summarise_sales(sales))
```

## Task 3: Loops

### a) Create a new list called `passing` that contains only the names of students with a grade of 50 or above. Print the list.

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

passing = []

# Loops through the students list and appends students with a grade of 50 or above to passing list
for i in students:
    name = i["name"]
    grade = i["grade"]
    
    if grade >= 50:
        passing.append(name)

print(f"List of passing students:\n{passing}")
```

### b) Loop through the students and print a status message for each. If the student's grade is 80 or above, print `"[name]: Excellent"`; if between 50 and 79, print `"[name]: Pass"`; otherwise print `"[name]: Fail"`.

```python
for i in students:
    name = i["name"]
    grade = i["grade"]

    # Checks in which range the students grade is and assigns the status message to status_message
    if grade >= 80:
        status_message = "Excellent"
    elif grade >= 50:
        status_message = "Pass"
    else:
        status_message = "Fail"
    
    print(f"{name}: {status_message}")
```

### c) Count how many students have an attendance rate below 0.80. Print the count.

```python
attendance_count = 0

# Loops through the students and counts the amount that have an attendance rate below 0.8
for i in students:
    attendance = i["attendance"]

    if attendance < 0.80:
        attendance_count += 1
    
print(f"Amount of students with an attendance rate below 0.80: {attendance_count}")
```

### d) Find the student with the highest grade using a loop (do not use `max()`). Print the student's name and grade.

```python
best_student = ""
best_student_grade = 0

# Loops through the students and assigns the name and grade of the student with the highest grade
for i in students:
    name = i["name"]
    grade = i["grade"]

    if grade > best_student_grade:
        best_student = name
        best_student_grade = grade

print(f"Student with the highest grade: {best_student} ({best_student_grade})")
```

### e) Now find the student with the highest grade using `max()`. Print the student's name and grade.

```python
# max() goes through each student in the list and for each student (x) use x["grade"] to compare grades
best = max(students, key = lambda x: x["grade"])

best_student = best["name"]
best_student_grade = best["grade"]

print(f"Student with the highest grade: {best_student} ({best_student_grade})")
```

## Task 4: Class and methods

```python
class Portfolio:
    def __init__(self, owner:str):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    def add_asset(self, ticker:str, shares:float, price_per_share:float):
        
        # Checks if there already exists shares with the same ticker and assigns the amount to current_shares
        current_shares = 0
        for key in self.holdings.keys():
            if key == ticker:
                current_shares = self.holdings[ticker]["shares"]

        shares_data = {}
        shares_data["shares"] = shares + current_shares # Adds the new amount of shares to the old ones
        shares_data["price_per_share"] = price_per_share

        self.holdings[ticker] = shares_data
    
    def total_value(self):
        self.total = 0
        for key, row in self.holdings.items():
            self.total += row["shares"] * row["price_per_share"]
        
        return round(self.total, 2)
    
    def most_valuable(self):
        highest_total_value = 0 # Temporary value
        highest_asset = "" # Temporary value
        
        # Loops through all keys and values, calculates a total value per asset and returns the ticker of the highest total value
        for key, row in self.holdings.items():
            row_value = row["shares"] * row["price_per_share"]
            if row_value > highest_total_value:
                highest_total_value = row_value
                highest_asset = key
        
        return highest_asset

    def __str__(self):
        assets_amount = len(self.holdings.keys())
        total = self.total_value()
        return f"Portfolio summary:\nOwner: {self.owner}\nNumber of assets: {assets_amount}\nTotal value: {total:,.2f}"

model_portfolio = Portfolio("Leander")

model_portfolio.add_asset("NVIDIA", 13, 175.20)
model_portfolio.add_asset("Nokia", 150, 7.28)
model_portfolio.add_asset("NVIDIA", 17, 172.37)
model_portfolio.add_asset("Apple", 15, 251.64)
```

## Task 5: Importing a module

```python
import datetime as dt

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

# Sets first_date and last_date to the first date in the orders list temporarily
first_date = dt.date.fromisoformat(orders[0][0])
last_date = dt.date.fromisoformat(orders[0][0])

# Assigns the first and last date to the actual first and last date in the list
for i in orders:
    original_date = i[0]
    date = dt.date.fromisoformat(original_date)
    if date < first_date:
        first_date = date
    if date > last_date:
        last_date = date

days_between_first_last_date = last_date - first_date

print(f"Amount of days between first and last date: {days_between_first_last_date.days}")
```

## Task 6: Array Operations and Broadcasting

### a) Compute and print the total annual revenue for each year (i.e., row sums). Use the `axis` parameter.

```python
import numpy as np

# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])

start_year = 2022

yearly_total = revenue.sum(axis = 1) * 1000 # Summarizes revenue per year and converts to euros

for i in range(len(yearly_total)):
    year = i + start_year
    
    year_revenue = yearly_total[i]

    print(f"{year}: {year_revenue:,.0f}€")
```

### b) Compute the mean quarterly revenue across all years (i.e., column means).

```python
quarter_mean = revenue.mean(axis = 0) * 1000

for index, i in enumerate(quarter_mean, 1):
    print(f"Q{index}: {i:,.0f}€")
```

### c) Using broadcasting, apply a growth adjustment to the 2024 row by multiplying it element-wise with the array `[1.05, 1.03, 1.08, 1.02]`. Print the adjusted values.

```python
growth_adjustment = [1.05, 1.03, 1.08, 1.02]

revenue_2024 = revenue[2]
adjusted_revenue_2024 = revenue_2024 * growth_adjustment * 1000

print("Adjusted revenue 2024:")
for quarter, value in enumerate(adjusted_revenue_2024, start = 1):
    print(f"Q{quarter}: {value:,.0f}€") # Prints the quarters one by one and displays in proper format
```

### d) Use boolean indexing to find and print all quarterly revenue values that exceed 250 (thousands of euros).

```python
print(revenue[revenue > 250])
```

## Task 8: More Pandas

### a) Create a DataFrame with columns: `date`, `visitors`, `conversion_rate`, and `avg_order_value`. Set `date` as the index.

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

# Creates a dictionary with the generated values
data = {
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value,
}

# Creates a DataFrame from the dictionary
website_traffic = pd.DataFrame(data, index= days)
website_traffic
```

### b) Compute two new columns: `conversions` (= visitors × conversion_rate, rounded to the nearest integer) and `daily_revenue` (= conversions × avg_order_value).

```python
# Creates column conversions and rounds to nearest integer
website_traffic["conversions"] = (website_traffic["visitors"] * website_traffic["conversion_rate"]).round(0).astype(int)

# Creates column daily_revenue (conversions * avg_order_value)
website_traffic["daily_revenue"] = website_traffic["conversions"] * website_traffic["avg_order_value"]

website_traffic
```
